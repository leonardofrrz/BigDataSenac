## transformando listas em pandas series

```python
import pandas as pd
exemplo={'num01' : 1,'num02' : 2,'num03' : 3,'num04' : 4,'num05' : 5}
exemplo=pd.Series(exemplo)
print('  Serie:')
print(exemplo)
```

```python
exemplo02=[1, 2, 3, 4, 5]
exemplo02=pd.Series(exemplo02, index = ['num01', 'num02', 'num03', 'num04', 'num05'])
print('  Serie:')
print(exemplo02)
```

```python
print('Prints\n',
    exemplo,'\n',
    exemplo02,'\n',
    exemplo02['num03'],'\n',
    exemplo02[4]
)
```
